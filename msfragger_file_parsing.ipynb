```python
def load_psm_df_msfragger(psm_file_path):
    psm_df = pd.read_table(psm_file_path)
    psm_df["temp_split_column"] = psm_df["Spectrum"].str.split(".")
    psm_df["Scan Number"] = psm_df["temp_split_column"].map(lambda x:x[1]).apply(pd.to_numeric)
    psm_df = psm_df.drop(columns=['Spectrum', 'temp_split_column','Mapped Genes', 'Mapped Proteins'])
    psm_df = psm_df.rename({'Protein ID': 'Protein Accession'}, axis=1)
    #psm_df = psm_df.set_index(['Spectrum File', 'Protein Accession', 'Peptide', 'Scan Number'])
    return psm_df
```

```python
# testing the loading function
psm_df = load_psm_df_msfragger(psm_file_path=psm_file_path)
psm_df
```

```python
peptide_file_path = "C:\\Users\\Sarah Curtis\\OneDrive - BYU\\Documents\\Single Cell Team Documents\\API_dev\\msfragger\\combined_peptide.tsv"

def load_pep_df_msfragger(peptide_file_path):
    pep_df = pd.read_table(peptide_file_path)
    pep_df = pep_df.rename({'Protein ID': 'Protein Accession', 'Peptide Sequence': 'Peptide'}, axis=1)
    return pep_df
```
